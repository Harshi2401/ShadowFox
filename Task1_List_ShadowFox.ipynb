```python
#Original Justice League list
justice_league = ["Superman","Batman","Wonder Woman","Flash","Aquaman","Green Lantern"]
print("original justice league:",justice_league)
```

```python
#Calculate number of members
print("number of members:",len(justice_league))
```

```python
#add Batgirl and Nightwing
justice_league.append("Batgirl")
justice_league.append("Nightwing")
print("After adding Batgirl and Nightwing:",justice_league)
```

```python
#Move Wonder Woman to the beginning (make her the leader)
#Move Wonder Woman to the beginning (make her the leader)
justice_league.remove("Wonder Woman")
justice_league.insert(0,"Wonder Woman")
print("Wonder woman is leader:",justice_league)
```

```python
#Move 'Green Lantern' between Aquaman and Flash
justice_league.remove("Green Lantern")
flash_index = justice_league.index("Flash")
justice_league.insert(flash_index,"Green Lantern")
print("After Reordering:",justice_league)
```

```python
#Crisis hits - Replace the entire team
justice_league=["Cyborg","Shazam","Hawkgirl","Martian Manhunter","Green Arrow"]
print("New Justice League:",justice_league)
```

```python
#Sort Alphabetically
justice_league.sort()
print("Sorted League:",justice_league)
print("Superhero New leader (first member):",justice_league[0])
```
